Data Collection, Analysis and Transformation

World Indices

```python
#loading the datasets to a pandas dataframe for World Indices

wi_df = pd.read_csv('/content/drive/MyDrive/Cohort4_Capstone_Project-1/World_Indices.csv')
```

```python
#printing the first 5 rows of the dataset

wi_df.head()
```

```python
#Check the datatypes

wi_df.dtypes
```

```python
#Check missing values

wi_df.isnull().sum()
```

```python
#Cleaning data

wi_df= wi_df.replace({',': '', '%': ''}, regex=True)
wi_df.head()
```

```python
#Convert into appropriate data types

columns_to_float = ['Last', 'High', 'Low', 'Chg.','Chg%']
wi_df[columns_to_float] = wi_df[columns_to_float].astype(float)
```

```python
from datetime import date
wi_df['Date/Time'] = pd.to_datetime(wi_df['Date/Time'],errors='coerce')
```

```python
# Convert values from ARS to USD for rows with 'Argentina' as the country (In dataset, the given corresponding values of Argentina are in ARS)

exchange_rate = 0.010
wi_df.loc[wi_df['Country'] == 'Argentina', ['Last', 'High', 'Low', 'Chg.']] *= exchange_rate

print(wi_df)
```

```python
#Check datatype

wi_df.dtypes
```

```python
#printing the first 5 rows of the dataset

wi_df.head()
```

```python
#Convert into CSV file

wi_df.to_csv("World_indices.csv", index=False)
```
